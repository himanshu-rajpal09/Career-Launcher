```python
df2['Day_Perc_Change'] = 100* df2['Close Price'].pct_change()
```

```python
df2['Trend'] = df2['Day_Perc_Change']
df2
```

```python
def raftaar(x):
    if x <= -7:
        return('Bear drop')
    elif (x >-7) & (x<-3):
        return('Among top losers')
    elif (x>= -3) & (x<=-1):
        return('Negative')
    elif (x>-1) & (x<-0.5):
        return('Slight negative')
    elif (x>= -0.5) & (x<=0.5):
        return('Slight or No change')
    elif (x>0.5) & (x<1):
        return('Slight positive')
    elif (x>=1) & (x<=3):
        return('Positive')
    elif (x>3) & (x<7):
        return('Among top gainers')
    elif (x>=7):
        return('Bull run')
    else: 
        pass
```

```python
df3 = df2.copy()
df3['Trend'] = df3['Trend'].apply(raftaar)
df3.head()
```

```python
df3.groupby('Trend').agg({'Total Traded Quantity':'mean'})
```
