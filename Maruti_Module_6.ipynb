```python
returns = df31.pct_change()*100
daily_return_avg = returns.mean()
daily_return_std = returns.std()
```

```python
annual_return = 252*daily_return_avg
volatility = np.sqrt(252)*daily_return_std
```

```python
df32 = pd.concat([annual_return,volatility], axis =1, keys = ['Return','Volatility'])
```

```python
df32
```

```python
from sklearn.cluster import KMeans
```

```python
K_clusters = range(1,10)
kmeans = [KMeans(n_clusters=i) for i in K_clusters]
Y_axis = df32[['Return']]
X_axis = df32[['Volatility']]
score = [kmeans[i].fit(Y_axis).score(Y_axis) for i in range(len(kmeans))]

# Visualize


plt.plot(K_clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init ='k-means++')

# Computing k-means clustering.

kmeans.fit(df32[df32.columns[0:2]]) # Computing k-means clustering.
df32['cluster_label'] = kmeans.fit_predict(df32[df32.columns[0:2]])

# Coordinates of cluster centers.

centers = kmeans.cluster_centers_ 

# Labels of each point

labels = kmeans.predict(df32[df32.columns[0:2]]) 
df32.head(30)
```

```python
plt.figure(figsize=(15,6))
df33 = df32[df32.cluster_label==0]
df34 = df32[df32.cluster_label==1]
df35 = df32[df32.cluster_label==2]
plt.scatter(df33.Return,df33['Volatility'],color='green', label = 'First_Group')
plt.scatter(df34.Return,df34['Volatility'],color='red', label = 'Second_Group')
plt.scatter(df35.Return,df35['Volatility'],color='black', label = 'Third_Group')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```
