```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df1 = pd.read_csv("C:\\Users\\Himanshu\\Downloads\\MARUTI.csv")
df1 = df1[df1.Series == 'EQ']
```

```python
df1['MA20'] = df1['Close Price'].rolling(window=20).mean()
df1['20dSTD'] = df1['Close Price'].rolling(window=20).std() 

df1['Upper'] = df1['MA20'] + (df1['20dSTD'] * 2)
df1['Lower'] = df1['MA20'] - (df1['20dSTD'] * 2)
df1['Middle'] =  (df1['Upper'] + df1['Lower'])/2 
```

```python
df1
```

```python
df1 = df1.loc[19:495,:].reset_index()
df1
```

```python
df1.drop(['Prev Close','Open Price','High Price','Low Price','Last Price'], axis = 1,inplace = True)
```

```python
df1.drop(['Total Traded Quantity','index','Series','Symbol','Date','Turnover','No. of Trades','Deliverable Qty','% Dly Qt to Traded Qty','MA20','20dSTD'], axis = 1, inplace =True)
```

```python
df1.drop('Average Price',axis = 1,inplace = True)
```

```python
def f(x):
    if x['Close Price'] < x['Lower']:
        return ('Buy')
    elif x['Close Price'] > x['Upper']:
        return ('Short')
    elif (x['Close Price']>x['Lower']) & (x['Close Price'] < x['Middle']):
        return('Hold Buy/Liquidate Short')
    elif (x['Close Price'] > x['Middle']) & (x['Close Price'] < x['Upper']):
        return('Hold Short/Liquidate Buy')
    else: 
        pass

df1['Call'] = df1.apply(f, axis=1)
```

```python
df1
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df1.drop(['Call','Close Price'],axis='columns'),df1.Call,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```

```python
df3 = pd.read_csv("C:\\Users\\Himanshu\\Downloads\\MARUTI.csv")
df3 = df3[df3.Series == 'EQ']
```

```python
df3.drop(['Total Traded Quantity','Series','Symbol','Date','Turnover','No. of Trades','Deliverable Qty','% Dly Qt to Traded Qty'], axis = 1, inplace =True)
```

```python
df3.head().reset_index()
```

```python
df3.drop(['Prev Close','Average Price','Last Price'], axis =1, inplace = True)
```

```python
df3.head()
```

```python
df3["Close_yesterday"] = df3["Close Price"].shift(1)
df3
```

```python
def apna(x):
    if x['Close Price'] < x['Close_yesterday']:
        return -1
    elif x['Close Price'] > x['Close_yesterday']:
        return 1
    else:
        pass
```

```python
df3['Action'] = df3.apply(apna, axis=1)
df3.drop('Close_yesterday', axis =1 ,inplace = True)
df3
```

```python
df3 = df3.loc[1:495,:]
df3
```

```python
df4 = df3.pct_change()*100
```

```python
df4['%CO'] = df4['Close Price'] - df4['Open Price']
df4['%HL'] = df4['High Price'] - df4['Low Price']
```

```python
df4.drop('Action', axis =1, inplace = True)
```

```python
df4 = df4.loc[1:495,:]
df4['MA5'] = df4['Close Price'].rolling(window=5).mean()
df4['5dSTD'] = df4['Close Price'].rolling(window=5).std()
```

```python
df4
```

```python
df5 = pd.concat([df3.Action, df4], axis = 1)
```
