```python
df5 = df5[df5.Series == 'EQ']
df6 = df6[df6.Series == 'EQ']
df7 = df7[df7.Series == 'EQ']
df8 = df8[df8.Series == 'EQ']
df9 = df9[df9.Series == 'EQ']
```

```python
df5 = df5.reset_index()
df6 = df6.reset_index()
df7 = df7.reset_index()
df8 = df8.reset_index()
df9 = df9.reset_index()
```

```python
df10 = pd.concat([df5['Close Price'], df6['Close Price'], df7['Close Price'], df8['Close Price'], df9['Close Price']], axis=1, keys=['Axis', 'Cipla','Titan','Reliance','Infy'])
```

```python
df10
```

```python
df10['Perc_Axis'] = 100* df10['Axis'].pct_change()
df10['Perc_Cipla'] = 100* df10['Cipla'].pct_change()
df10['Perc_Titan'] = 100* df10['Titan'].pct_change()
df10['Perc_Reliance'] = 100* df10['Reliance'].pct_change()
df10['Perc_Infy'] = 100* df10['Infy'].pct_change()
```

```python
df11 = df10.loc[:,'Perc_Axis':'Perc_Infy']
```

```python
#2.10

rolling_mean = df4['Close Price'].rolling(14).mean()
rolling_std = df4['Close Price'].rolling(14).std()
bollinger_upper_band = rolling_mean + rolling_std * 2
bollinger_lower_band = rolling_mean - rolling_std * 2
```

```python
#Visualize the closing price history
plt.figure(figsize=(20,10))
plt.title('Maruti Bollinger Bands')
ax = df4['Close Price'].plot(title="Bollinger Bands", fontsize=12)
rolling_average = rolling_mean
rolling_average.plot(label='Rolling mean', ax=ax)
bollinger_upper_band.plot(label='Bollinger upper band', ax=ax)
bollinger_lower_band.plot(label='Bollinger lower band', ax=ax)
plt.show()
```
